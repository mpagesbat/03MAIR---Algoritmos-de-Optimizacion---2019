**(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta Argumenta porque crees que mejora el algoritmo por fuerza bruta**

Este algoritmo mejora la fuerza bruta, gracias al hecho que cortamos ciertas ramas que sabemos que no llevan al resultado óptimo. <br>
Seria un algoritmo estilo Backtracking por etapas que cuando se encuentra que está por un camino que no lleva a una solución, retrocede para ir a por el siguiente valor. Hay varias etapas:

etapa 1: tenemos CIFRA+SIMBOLO+CIFRA

etapa 2: tenemos la etapa 1 + SIMBOLO + CIFRA. en esta etapa hacemos la primera poda de valores según propiedades matemáticas de cálculo que aseguran que no se puede conseguir el resultado deseado.

etapa 3: tenemos la etapa 3 + SIMBOLO + CIFRA. en esta etapa hacemos la segunda poda de valores según propiedades matemáticas de cálculo que aseguran que no se puede conseguir el resultado deseado.

Etapa 4:etapa 3 + SIMBOLO + CIFRA calculamos el valor final, y comparamos con el deseado.

-las ramas que cortamos son: <br>
     

*   en la etapa 2 cuando tenemos cifra+simbolo+cifra+simbolo+cifra, si los dos simbolos son / y *, y el resultado es decimal, podemos asegurar que el resultado final no serà entero.

*   en la etapa 2 si solo hemos usado / y *, si el valor es mas de 9ud más grande que el valor deseado, o 9ud mas pequeño que el valor deseado, serà imposible llegar al valor deseado.

*   en la etapa 2 si solo hemos usado / y *, podemos asegurar que no obtenemos el valor final.

*   en la etapa 3, si solo nos queda por poner un + o - y el resultado provisional es un decimal, cortamos porque nunca podemos obtener un resultado entero

*   en la etapa 3, si el valor provisional es igual al valor deseado, es imposible obtener con la ultima iteración el resultado deseado y por lo tanto cortamos

Después de estas restricciones, el algoritmo funciona mucho más eficiente y rápido en calcular.


```
#####ALGORITMO CON COMENTARIOS #####
import math
string=""
lista1=["*","+","-","/"]   #lista de signos
lista2=["1","2","3","4","5","6","7","8","9"] #lista de valores.

y=43

def evaluacion_final(a,b,c,d,e,f,g,h,i,y):
  string=""
  string=(a+b+c+d+e+f+g+h+i)  #creamos el string con todas las cifras y simbolos
  q=eval(string) #evaluamos el string para tener el calculo del valor
  partedecimal, parteentera = math.modf(q)  #miramos si tiene decimal el valor
  if partedecimal==0:  #si no tiene decimal, puede seguir
    if q==y:  #si es igual que el valor deseado, es solucion final.
      print("buen resultado",string,"igual a ",q)
  string=""

def evaluacion_tercera(a,b,c,d,e,f,g,lista7,lista8):  #evaluacion en la 3a etapa
  string2=""
  string2=(a+b+c+d+e+f+g) #creamos el string para calcular                    
  q=eval(string2)
  partedecimal, parteentera = math.modf(q)
  if partedecimal!=0 and ("+" in lista8 or "-" in lista8): return False  #si hay decimales, y la ultima fase es un + o - seguro que no es solucion 
  if q==y: return False   #si tenemos el valor final, en esta etapa no podemos tener solucion 
  return True   

def evaluacion_segunda(a,b,c,d,e,lista5,lista6):
  string3=""
  string3=(a+b+c+d+e)                       
  q=eval(string3)
  partedecimal, parteentera = math.modf(q)
  if partedecimal!=0 and ("-" in lista6 and "+" in lista6): return False  #si el numero es decimal y falta por poner + y -, no se llega a solución 
  if  ("+" in lista6 and "-" in lista6)and (q>y+9 or q<y-9) : return False #si el valor es mas grande que y+9 o mas pequeño que y-9, no solución.
  if q==y and ("-" in lista6 and "+" in lista6): return False  #si tenemos el valor final, en esta etapa no podemos tener solucion
  return True   

for a in lista2:  #etapa 1
  for b in lista1:
    lista3=lista2[:] #copiamos la lista para la siguente fase
    lista3.remove(a) #eliminamos una cifra de la lista de cifras.
    lista4=lista1[:]
    lista4.remove(b) 
    for c in lista3:
      for d in lista4:
        lista5=lista3[:]
        lista5.remove(c)
        lista6=lista4[:]
        lista6.remove(d)
        for e in lista5:#etapa 2
          if evaluacion_segunda(a,b,c,d,e,lista5,lista6): #evaluamos los valores que no pasan a la siguiente etapa 
            for f in lista6:
              lista7=lista5[:]
              lista7.remove(e)
              lista8=lista6[:]
              lista8.remove(f)
              for g in lista7:#etapa 3 
                if evaluacion_tercera(a,b,c,d,e,f,g,lista7,lista8): #evaluamos los valores que no pasan a la siguiente etapa 
                  for h in lista8:
                    lista9=lista7[:]
                    lista9.remove(g)      
                    for i in lista9:
                      evaluacion_final(a,b,c,d,e,f,g,h,i,y) #evaluacion final de si el valor es el deseado.       
              lista8.append(g) #a partir de aqui añadimos los valores a la lista otra vez.
              lista7.append(f)
        lista6.append(e)
        lista5.append(d)
    lista4.append(c)
    lista3.append(b)

```
